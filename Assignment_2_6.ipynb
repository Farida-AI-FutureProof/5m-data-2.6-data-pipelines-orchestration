```python
# ============================================
# Cell 2: Initialise a new Meltano project
# ============================================
import os

# Create a working directory for the demo
project_root = "/content/meltano_pg_bq_demo"
os.makedirs(project_root, exist_ok=True)

%cd $project_root

# Initialise Meltano project
!meltano init elt_project

# Move into the project directory
%cd elt_project

# Show project structure (optional, but nice for markers)
!ls
```

```python
# ============================================
# Cell 3: Add PostgreSQL tap and BigQuery target
# ============================================

# Add the PostgreSQL extractor
!meltano add extractor tap-postgres

# Add the BigQuery loader
!meltano add loader target-bigquery

# Show plugins now registered in meltano.yml
!grep -A5 "plugins:" -n meltano.yml || cat meltano.yml
```
